```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
df=pd.read_csv('LS2014Electors.csv')
```

```python
df1=df[df['POLL PERCENTAGE']>85]
```

```python
#constituencies with poll percentage greater than 85%
plt.figure(figsize=(20,10))
plt.xlabel("Constituencies")
plt.ylabel("Poll Percentage")
plt.barh(df1['PARLIAMENTARY CONSTITUENCY'],df1['POLL PERCENTAGE'])
```

```python
#Pie chart on Poll percentage Bengaluru 
df2=df[(df['STATE']=='Karnataka') & (df['PC NO']>22) & (df['PC NO']<27)]
```

```python
plt.figure(figsize=(20,10))
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0, 0, 0, 0.1)
plt.pie(df2['Total voters'], explode=explode, labels=df2['PARLIAMENTARY CONSTITUENCY'], colors=colors,autopct='%1.1f%%', shadow=True, startangle=140)
#plt.axis('equal')
plt.title("Total Voters:5002124")
plt.show()
```
